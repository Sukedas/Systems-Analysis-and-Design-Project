# Cellular Automata Simulation
## Microenterprise Density Clustering

Simulating the growth and decay of density clusters on a 50x50 grid.

```python
import sys
sys.path.append('../src')
from cellular_automata import MicroEnterpriseCA
import matplotlib.pyplot as plt

ca = MicroEnterpriseCA(grid_size=50)
ca.initialize_random(density=0.2)

history = ca.run_simulation(steps=20)

# Visualize final step
plt.imshow(ca.grid, cmap='Greens')
plt.title('Final State')
plt.show()
```
