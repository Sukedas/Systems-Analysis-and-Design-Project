# Exploratory Data Analysis (EDA)
## Microenterprise Density Prediction

This notebook explores the dataset, checking for trends, seasonality, and outliers.

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
from utils import load_config

config = load_config()
train_df = pd.read_csv(f"{config['DATA_PATH']}/train.csv")
train_df.head()
```

```python
# Plot Density over Time for a sample county
sample_cfips = train_df['cfips'].unique()[0]
subset = train_df[train_df['cfips'] == sample_cfips]
plt.plot(pd.to_datetime(subset['first_day_of_month']), subset['microbusiness_density'])
plt.title(f'Density for CFIPS {sample_cfips}')
plt.show()
```
